# **Control Flow**

**Control Flow** means the order in which the computer executes instructions in a program.
By default, code runs from top to bottom.

- But sometimes, we want the program to make decisions:
    - If something is true â†’ do this
    - Otherwise â†’ do something else

> Control flow allows programs to choose different paths depending on conditions.

## **if â€“ elif â€“ else Statement**

```python
age = 18

if age < 13:
    print("You are a child")
elif age < 18:
    print("You are a teenager")
else:
    print("You are an adult")
```

## **Successive if Statements (Multiple Independent If)**

```python
temperature = 30

if temperature > 20:
    print("It's warm")

if temperature > 25:
    print("It's hot")

if temperature > 30:
    print("It's very hot")
```

> An **ifâ€“elifâ€“else** chain checks conditions in order and stops as soon as one condition is true, meaning only one block can run; in contrast, using several separate **if statements** means each condition is tested independently, so multiple blocks can run if several conditions are true.

## **Nested if Statement**

```python
age = 20
has_id = True

if age >= 18:
    if has_id:
        print("You can enter")
    else:
        print("You need an ID")
else:
    print("You are too young")
```

## **logical operators**

```python
age = 25
has_ticket = True
is_vip = False

if age >= 18 and has_ticket and not is_vip:
    print("You can enter the regular area.")
elif age >= 18 and (has_ticket or is_vip):
    print("You can enter either VIP or regular area.")
elif not has_ticket:
    print("You cannot enter without a ticket.")
else:
    print("You do not meet the entry requirements.")
```

## **Lists**

In Python, a **List** is a built-in data structure used to store a collection of items within a single variable. It is one of the most versatile and commonly used tools for organizing data.

> You can think of a list as a container that holds a sequence of elements, keeping them organized in a specific way.

**Key Properties**

> Here are the four fundamental characteristics that define how a list behaves:
1. **Ordered** When you add items to a list, they maintain a specific order. The first item you put in stays at the beginning, and the last item stays at the end (unless you explicitly move them). Each item has a designated position number (called an index), allowing you to find an item based on where it sits in the sequence.
2. **Mutable (Changeable)** Lists are dynamic. After you create a list, you are free to modify it. You can add new items, delete existing items, or replace an item with something else. The list can grow or shrink in size automatically as you modify it.
3. Allows **Duplicates** A list does not require its items to be unique. You can store the same value multiple times in the same list, and each instance will be treated as a separate item with its own position.
4. **Heterogeneous (Mixed Data Types)** A list can hold different types of data at the same time. You do not need to restrict it to just numbers or just text. A single list can contain an integer, a decimal number, a string of text, and even another list all mixed together.

## **Dictionaries**

**Dictionaries** in Python are best understood as unordered collections of data that store information in a key-value pairing structure.

Here's a breakdown of their key features:
- **Mapping**: They represent a mapping between unique keys and their associated values. Think of it like a physical dictionary or a phone book: the word (key) is mapped to its definition (value), or a person's name (key) is mapped to their phone number (value).
- **Keys**: The keys must be unique and immutable (cannot be changed), such as strings, numbers, or tuples. They are used to quickly look up and retrieve the corresponding value.
- **Values**: The values can be of any data type and can be duplicated. They are the actual data you want to store.
- **Mutability**: The dictionary itself is mutable, meaning you can add new key-value pairs, change the values associated with existing keys, or remove pairs after the dictionary has been created.
- **Unordered (in older Python versions)**: Traditionally, the order in which items were inserted was not guaranteed. However, since Python version 3.7 (and 3.6 for CPython), dictionaries maintain the insertion order of their keys.

```python
language = {
    "name": "Python",
    "creator": "Guido van Rossum",
    "year": 1991,
    "is_dynamic": True,
    "versions": ["2.x", "3.x"]
}

print(language)
print("Language Name: " + language["name"])
print("Creator: " + language["creator"])
print("Year Created: " + str(language["year"]))
print("Is Dynamic Typing: " + str(language["is_dynamic"]))
print("Versions: " + ", ".join(language["versions"]))
```

# **Local vs Global Scope**

**ðŸŒ Global Scope**

The global scope refers to the entire program.
- **Definition**: Any variable defined outside of all functions, classes, or blocks (like loops or conditional statements) is in the global scope.
- **Accessibility**: Variables in the global scope are accessible and can be read from anywhere within the entire program, including inside any function or class.
- **Lifetime**: They exist from the moment they are defined until the program finishes executing.
- **Modification**: To change a global variable's value from inside a function, you must explicitly declare it as global within that function first. If you don't, Python assumes you are creating a new local variable with the same name.

**ðŸ  Local Scope**

The local scope refers to a limited section of the program, such as inside a function.
- **Definition**: A variable is in a local scope if it is defined inside a function (or a class method).
- **Accessibility**: Local variables are only accessible and can only be used from within the function where they are defined. Once the function finishes executing, the local variables are destroyed.
- **Lifetime**: They exist only for the duration of the function call.
- **Isolation**: A function's local scope is isolated from other functions' local scopes and the global scope. This means different functions can safely use the same variable name without their values interfering with each other.

```python
# Global variable
player_healthy = 10

def game():
    # Local variable
    player_healthy = 6
    print("Player healthy inside function is : " + str(player_healthy))

game()
print("Player healthy outside function is : " + str(player_healthy))
```

## **Block Scope**

In Python, the concept of **"block scope"** as it is typically defined in languages like **C++**, **C#**, or **Java** (where a variable's visibility is limited to an arbitrary block of code enclosed by curly braces, e.g., an **if** statement or a **while** loop) does not exist.

Instead, Python primarily uses function scope and module scope.

Here is a definition focusing on what Python uses instead of true block scope:
- **Block Scope** (in the context of Python's scoping rules):
- **In Python**, a code block created by an if statement, a while loop, or a for loop does not create a new, separate scope that isolates variables defined inside it.
- **Variables** defined within these blocks are still considered part of the enclosing function's scope (or the global/module scope if they are defined at the top level). This means that once the block finishes executing, the variables remain accessible and mutable outside of that specific block.
    - The primary scope boundaries that do exist in Python are:
        - **Function/Method** Definitions (def)
        - **Classes** (class)
        - **Modules** (the Python file itself)

```python
def scope_test():
    # 1. Start of the function scope
    message = "Hello from the function scope_test!"
    
    if True:
        # This is a block, but NOT a new scope boundary in Python
        message = "Hello from inside the IF block!"
        # 'message' is created in the function's scope

    # The block has ended, but 'message' is still accessible here.
    print(f"Inside the function, after the IF: {message}")

scope_test()

# Output: Inside the function, after the IF: Hello from inside the IF block!
```

### **Modifying Global Variables**

**âŒ Modifying Global Variables: Generally Discouraged**

In most cases, and especially in large or complex programs, it is not considered a good practice to directly modify global variables from within functions or other parts of your code.

While Python allows you to do this using the global keyword, most developers try to minimize its use.

Here's a breakdown of why, and what the better alternatives are:
**ðŸš¨ Why Modifying Globals is Problematic**
1. **Difficult to Debug (Hidden Dependency):** When a function modifies a global variable, the function's output or effect is no longer solely determined by the arguments passed to it. If the global variable's value is wrong, it can be extremely difficult to trace which function changed it and when. This is often called hidden dependency or side effect.
2. **Reduced Readability:** A function's purpose should be clear just by looking at its definition and arguments. When a function modifies a global, you have to look outside the function to understand its complete behavior.
3. **Breaks Modularity and Reusability:** Functions that rely on or modify globals are tightly coupled to that specific global variable. You cannot easily reuse that function in a different part of the program or in a new project without bringing the global variable along.
4. **Concurrency Issues (In Multithreading):** If multiple threads or processes try to modify the same global variable at the same time, it can lead to unpredictable and potentially disastrous race conditions, where the final value depends on which thread finishes last.

âœ… Better Alternatives to Global Modification
Instead of modifying a global variable, the preferred patterns are:

Alternative______________Description_______________Example (Conceptual)
1. **Pass and Return :** "Pass the value of the variable to the function, and have the function return the modified value. This is the most common and cleanest solution. => **new_data = modify(old_data)**
2. **Use Classes :** "Use Object-Oriented Programming (OOP) where the ""global"" data is stored as an instance variable (attribute) of a class, and functions (methods) act only upon that instance." => **my_object.update_value(new_input)**
3. **Use Constants :** "If the global variable is a value that should never change (like a configuration setting or Ï€), define it once in capital letters (e.g., MAX_ATTEMPTS = 5) and use it read-only. This is the safest use of a global variable." => **if attempt < MAX_ATTEMPTS:**

```python
## ðŸ›‘ THE GLOBAL VARIABLE APPROACH (Less Preferred)

# 1. Define the Global variable at the top level
USER_SCORE = 100 

def update_global_score(points_earned):
    # To modify the global variable, we must use the 'global' keyword.
    # This creates a hidden side effect outside the function's visible inputs/outputs.
    global USER_SCORE 
    USER_SCORE += points_earned
    # Note: This function doesn't return anything useful, it just changes a global state.

print(f"Initial Global Score: {USER_SCORE}")
update_global_score(50) 
print(f"Score AFTER global modification: {USER_SCORE}") 

print("-" * 30)

## âœ… THE PASS AND RETURN APPROACH (Preferred)

# 1. Define the initial variable (can be global or local)
current_hp = 200

def calculate_new_health(hp, damage_taken):
    # The function ONLY works with its local inputs (hp, damage_taken).
    # It has NO hidden dependencies on anything outside its arguments.
    new_hp = hp - damage_taken
    
    # It returns the result, making its effect clear and predictable.
    return new_hp

print(f"Initial Local Health: {current_hp}")

# 2. Call the function and explicitly assign the returned value back.
new_health = calculate_new_health(current_hp, 35)

print(f"Health AFTER calculating new value: {new_health}")
# The original variable 'current_hp' remains unchanged unless explicitly reassigned.
print(f"Original 'current_hp' variable (still 200): {current_hp}")
```

### **Constants**

**ðŸ’¡ Constants in Python**

In **Python**, the concept of a "constant"â€”a variable whose value is fixed and cannot be changed after it is assignedâ€”does not exist at the language level.

Unlike languages like C++ or Java, Python does not have a built-in keyword (like const or final) that prevents a variable from being reassigned or modified.

Instead, Python relies on convention and programmer discipline to designate and treat variables as constants.

1. **The Convention: Naming**

To signal to all other programmers (and your future self) that a variable is intended to be a constant, you use a specific naming convention:
- ALL CAPS with Underscores: The variable name should be written entirely in uppercase letters, using underscores (_) to separate words.

- Purpose__________________Example
- Configuration Setting____MAX_CONNECTIONS = 100
- Mathematical_____________Value,PI = 3.14159
- Status Indicator_________DEFAULT_TIMEOUT = 30

2. **The Implementation: A Warning, Not a Barrier**

When you define a variable using the all-caps convention, you are essentially issuing a warning: "Do not reassign this value."
- **You can still change it:** If another part of the code ignores the convention and tries to change the value, Python will not throw an error. The variable will simply be reassigned to the new value.
- **Best Practice:** Good Python programmers respect this convention. If they see an ALL_CAPS variable, they know they should only use its value and never attempt to modify it.

3. **Where to Define Constants**

The most common and recommended practice is to define all constants in a separate file, typically named **constants.py** or **config.py**.
This centralizes all the immutable configuration and magic numbers, making them easy to find, import, and manage across your entire project.
